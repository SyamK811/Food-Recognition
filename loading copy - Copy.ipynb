Import Test Data

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model,Model

test_data_dir = r"C:\Users\HP-NBT\Desktop\project\food101\test"
test_data = tf.keras.preprocessing.image_dataset_from_directory(test_data_dir)
```

```python
height = 299
width = 299
channels = 3
batch_size = 32
img_shape = (height, width, channels)
img_size = (height, width)

datagen = ImageDataGenerator(
        rescale = 1./255,
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2, 
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
)

test_data = datagen.flow_from_directory(
    test_data_dir,
    target_size = img_size,
    batch_size = batch_size,
    class_mode='categorical',
    shuffle=False)
```

Evaluating the model

```python
test_loss, test_accuracy = model.evaluate(test_data)
print(f"Test Accuracy: {test_accuracy}")
print(f"Test Loss: {test_loss}")
```

Confusion Matrix

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

#Get model predictions on the test data
predictions = model.predict(test_data)

# If your model outputs probabilities, convert them to class labels (i.e., highest probability)
predicted_labels = np.argmax(predictions, axis=1)

#Extract true labels from the test data
true_labels = test_data.labels  

#ompute the confusion matrix
cm = confusion_matrix(true_labels, predicted_labels)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=np.arange(cm.shape[0]), yticklabels=np.arange(cm.shape[0]))
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Assuming your test data is a generator like ImageDataGenerator
class_names = list(test_data.class_indices.keys())

# Step 5: Slice the confusion matrix to keep only the first 10 categories
cm_20 = cm[:20, :20]  # Get the first 10 rows and columns

# Step 6: Visualize the confusion matrix for the first 10 categories with class names
plt.figure(figsize=(16, 12))
sns.heatmap(cm_20, annot=True, fmt="d", cmap="Blues", 
            xticklabels=class_names[:20], yticklabels=class_names[:20])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix (First 20 Categories)')
plt.show()
```

Food Prediction

```python
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# Define the function to predict the class of an image
def predict_image(filename, model, classes):
    # Load the image from the file and resize it to 299x299
    img_ = image.load_img(filename, target_size=(299, 299))  
    img_array = image.img_to_array(img_)  # Convert image to NumPy array
    img_processed = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_processed = img_processed.astype('float32') / 255.0  # Normalize (scale between 0 and 1)

    # Predict the class of the image using the provided model
    prediction = model.predict(img_processed)

    # Get the index of the highest predicted class
    index = np.argmax(prediction)
    
    # Show the image and display the predicted class
    plt.title(f"Prediction - {str(classes[index]).title()}", size=15, color='red')
    plt.imshow(img_array.astype('uint8'))  # Display the original image
    plt.axis('off')  # Hide axes for better viewing
    plt.show()


    # Return the predicted class
    return classes[index]
```
